```python
# Q 71. Implement recursive Linear Search
def linear_search_r(arr, size, val):
  if size == 0:
    return -1

  if arr[size - 1] == val:
    return size - 1
  else:
    return linear_search_r(arr, size - 1, val)


a = [1, 2, 0, 3, 9, 4, 7]
size = len(a)
val = 3
print(linear_search_r(a, size, val))     # Output: 3
```

```python
# Q 72. Implement Iterative Linear Search
def linear_search_i(arr, size, val):
  for i in range(size):
    if arr[i] == val:
      return i

  return -1


a = [3, 7, 2, 4, 8, 10, 5]
size = len(a)
val = 5
print(linear_search_i(a, size, val))      # Output: 6
```

```python
# Q 73. Implement Recursive Binary Search
def binary_search_r(arr, val, low, high):
  if low > high:
    return -1
  else:
    mid = (low + high) // 2
    if arr[mid] == val:
      return mid
    elif arr[mid] > val:
      return binary_search_r(arr, val, low, mid - 1)
    else:
      return binary_search_r(arr, val, mid + 1, high)


a = [2, 5, 7, 18, 20, 30]
low = 0
high = len(a) - 1
val = 20
print(binary_search_r(a, val, low, high))
```

```python
# Q 74. Implement Iterative Binary Search
def binary_search_i(arr, val, low, high):
  while low != high:
    mid = (low + high) // 2
    if arr[mid] == val:
      return mid
    elif arr[mid] > val:
      high = mid - 1
    else:
      low = mid + 1

  return -1


a = [2, 5, 7, 18, 20, 30]
low = 0
high = len(a) - 1
val = 20
print(binary_search_i(a, val, low, high))
```

```python
# Q 75. Implement Bubble Sort
def bubble_sort(arr, size):
  for i in range(size):
    for j in range(size - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]

  return arr


a = [3, 5, 2, 10, 4, 1, 7]
size = len(a)
print(bubble_sort(a, size))
```

```python
# Q 76. Implement Selection Sort
def selection_sort(arr, size):
  for i in range(size):
    min_ = i
    for j in range(i, size):
      if arr[j] < arr[min_]:
        min_ = j

    if min_ != i:
      arr[i], arr[min_] = arr[min_], arr[i]

  return arr


a = [3, 5, 2, 10, 4, 1, 7]
size = len(a)
print(bubble_sort(a, size))
```

```python
# Q 77. Implement Matrix multiplication
def matrix_mul(A, B):
  if len(A[0]) != len(B):
    raise Exception(
        f"Invalid Matrix sizes for Multiplication : {len(A)} x {len(A[0])} and {len(B)} x {len(B[0])} -> {len(A[0])} != {len(B)}")
  else:
    m1, n1 = len(A), len(A[0])     # Matrix 'A' is of order m1 x n1
    m2, n2 = len(B), len(B[0])     # Matrix 'B' is of order m2 x n2
    # The reslut of A x B is the matrix of order m1 x n2. List comprehensions used instead of [...] * n to prevent aliasing
    res = [[0 for _ in range(n2)] for _ in range(m1)]

    for i in range(m1):
      for j in range(n2):
        s = 0
        for k in range(n1):
          x = A[i][k]
          y = B[k][j]
          s += x * y
        res[i][j] = s

    return res


a = [[1, 2, 3], [2, 3, 4], [3, 4, 5], [4, 5, 6], [5, 6, 7]]  # Order 5 x 3
b = [[3, 2, 1, 4], [0, 3, 2, 5], [1, 5, 5, 7]]              # Order 3 x 4
# b = [[3, 2, 1], [0, 3, 2]]                                # Order 2 x 3 : Invalid Matrix sizes for Multiplication
print(matrix_mul(a, b))
```

```python
# Q 78. Implement Matrix Addition
def matrix_add(A, B):
  if len(A) != len(B) or len(A[0]) != len(B[0]):
    raise Exception(
        f"Invalid Matrix sizes for Addition : {len(A)} x {len(A[0])} and {len(B)} x {len(B[0])}")
  else:
    res = [[0 for _ in range(len(A[0]))] for _ in range(len(A))]
    for i in range(len(A)):
      for j in range(len(A[0])):
        res[i][j] = A[i][j] + B[i][j]

    return res


a = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]   # Order 3 x 3
b = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]   # Order 3 x 3
# b = [[1, 2, 3], [2, 3, 5]]            # Order 2 x 3 : Invalid Matrix sizes for Addition
print(matrix_add(a, b))
```

```python
# Q 89. Implement stack and basic operations

# A stack follows LIFO (Last In, First Out). So elements are always added to the end (called push) and are removed starting from end (called pop)
stack = [1, 2, 3, 4]
print("Initial Stack\n", stack, "\n")

# Adding Elements
stack.append(7)
stack.append(9)
stack.append(5)
print("After pushing elements\n", stack, "\n")

# Removing Elements
print("Removing Elements")
print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())
print("\nAfter popping elements\n", stack)

# isFull and isEmpty can be used to check the status of stack
# let the max size of stack be 5
print("isFull:", len(stack) == 5)
print("isEmpty:", len(stack) == 0)

# top can be used to get the top element of stack
print("Top:", stack[-1])
```

```python

# Q 90. Implement queue and basic operations
# A Queue follows FIFO (First In, First Out). Elements are removed in the same order as they were added
queue = []
print("Initial Queue\n", queue, "\n")

# Adding elements to the Queue is called Enqueuing (Enqueue)
queue.append("a")
queue.append("b")
queue.append("c")
queue.append("d")
print("After Enqueuing\n", queue, "\n")

# Removing elements from Queue is called Dequeuing (Dequeue)
print("Dequeuing")
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0), "\n")

print("After Dequeuing\n", queue, "\n")

# isFull and isEmpty can be used to check if the Queue is full or empty respectively
# let the maximum size of the Queue be 4
print("isFull:", len(queue) == 4)
print("isEmpty:", len(queue) == 0)

# rear and front can be used to get the last and first elements of the Queue respectively
print("rear:", queue[-1])
print("front:", queue[0])
```

```python
# Q 91. Implement Linked List and basic operations (Using List)
# A Linked List is a linear data structure where each element is a separate object. 
# Each element (we will call it a node) of a list is comprising of two items - the data and a reference to the next node.

# Singly linked list with Python list
ll = [1, 2, 3, 4, 5]
print("Initial Linked List\n", ll, "\n")
print("IDs of nodes:", *[id(i) for i in ll], "\n")
print("Length of Linked List:", len(ll), "\n")

# Adding elements to the Linked List is called Insertion
ll.append(6)
ll.append(7)
ll.append(8)
print("After Insertion\n", ll, "\n")
print("IDs of nodes:", *[id(i) for i in ll], "\n")
print("Length of Linked List:", len(ll), "\n")

# Removing elements from the Linked List is called Deletion
print("Deletion")
print(ll.pop(0))
print(ll.pop(0))
print(ll.pop(0), "\n")

print("After Deletion\n", ll, "\n")
print("IDs of nodes:", *[id(i) for i in ll], "\n")
print("Length of Linked List:", len(ll), "\n")

# Reverse a Linked List
print("Reversing Linked List")
ll.reverse()
print(ll, "\n")

# Sorting a Linked List
print("Sorting Linked List")
ll.sort()
print(ll, "\n")
```

```python
# Q 91. Implement linked list and basic operations (Using Classes)
# A linked list contains elements which are not stored in contiguous memory locations.
# They contain a reference to the next element in the list

# Implementing a Singly Linked List
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

  def __str__(self):
    return f"Node<data={self.data}, next={'Node<...>' if self.next is not None else None}>"


class LinkedList:
  def __init__(self):
    self.head = None

  # Displaying the LinkedList
  def __repr__(self):
    temp = self.head
    text = "LinkedList["
    while temp is not None:
      text += str(temp.data)
      if temp.next is not None:
        text += " -> "
      temp = temp.next
    text += "]"
    return text

  def __str__(self):
    temp = self.head
    text = "["
    while temp is not None:
      text += str(temp.data)
      if temp.next is not None:
        text += " -> "
      temp = temp.next
    text += "]"
    return text

  # Add a new item at front
  def insert_at_head(self, val):
    new_head = Node(val)
    new_head.next = self.head
    self.head = new_head
    return new_head

  # Add a new item after specified node
  def insert_after(self, prev_node, val):
    new_item = Node(val)
    new_item.next = prev_node.next
    prev_node.next = new_item
    return new_item

  # Add a new item at end
  def append(self, val):
    tail = Node(val)
    last = self.head
    if last is None:
      self.head = tail
    else:
      while last.next is not None:
        last = last.next
      last.next = tail
    return tail

  # Searching through the list
  def search(self, val):
    pos = 0
    temp = self.head
    while temp is not None:
      if temp.data == val:
        return temp, pos
      temp = temp.next
      pos += 1

  # Removing head
  def remove_head(self):
    if not self.head:
      return
    new_head = self.head.next
    self.head = new_head

  # Removing at given position (similar to index)
  def remove_at(self, pos):
    if pos == 0:
      return self.remove_head()
    prev = self.head
    for i in range(pos - 1):
      if prev == None:
        return
      prev = prev.next
    next = prev.next.next if prev.next is not None else None
    prev.next = next

  def remove_node(self, val):
    temp = self.head
    if temp.data == val:
      return self.remove_head()
    while temp.next is not None:
      if temp.data == val:
        prev.next = temp.next
        break
      prev = temp
      temp = temp.next

  def remove_after(self, prev_node):
    to_remove = prev_node.next
    if to_remove is None:
      return
    after = to_remove.next
    prev_node.next = after

  def remove_tail(self):
    if self.head is None:
      return
    if self.head.next is None:
      self.head = None
      return
    prev = self.head
    while prev.next.next is not None:
      prev = prev.next
    prev.next = None

  def reverse(self):
    prev = None
    curr = self.head
    while curr is not None:
      next = curr.next
      curr.next = prev
      prev = curr
      curr = next
    self.head = prev

  def sort(self):
    if self.head is None:
      return
    curr = self.head
    while curr.next is not None:
      next = curr.next
      while next is not None:
        if curr.data > next.data:
          curr.data, next.data = next.data, curr.data
        next = next.next
      curr = curr.next

  def __len__(self):
    temp = self.head
    count = 0
    while temp is not None:
      count += 1
      temp = temp.next
    return count



# Creating a new LinkedList
l_list = LinkedList()
print("Instantiate an empty linked list")
print(l_list)

# Adding elements
# i. To end
l_list.append(2)
print("\nAppending a node with value 2")
print(l_list)

# ii. At Head
print("\nInserting at head a node with value 7")
prev_node = l_list.insert_at_head(7)
print(l_list)

# iii. After specified node
print("\nInserting after a given node")
l_list.insert_after(prev_node, 3)
l_list.insert_after(prev_node, 11)
l_list.insert_after(prev_node, 4)
print(l_list)

# Removing elements
# i. Removing head
print("\nRemoving Head")
l_list.remove_head()
print(l_list)

# ii. Removing a node with specified value
print("\nRemoving node with specified value")
l_list.remove_node(3)
print(l_list)

# iii. Removing a node at specified position
print("\nRemoving node at specified position")
l_list.remove_at(1)
print(l_list)

head = l_list.head

# iv. Removing a node after specified node
print("\nRemoving node after a specific node")
l_list.remove_after(head)
print(l_list)

# Repopulating the linked list
print("\nRepopulating list to experiment more commands")
l_list.append(1)
l_list.append(4)
l_list.append(9)
l_list.append(16)
print(l_list)

# Searching for an item by value
print("\nSearching for a node by value")
node, pos = l_list.search(4)
print(node, "at pos.", pos)

# Reversing the linked list
print("\nReversing the linked list")
l_list.reverse()
print(l_list)

# Sorting the linked list
print("\nSorting the linked list")
l_list.sort()
print(l_list)

# Length of the linked list
print("\nLength of the linked list")
print(len(l_list))

```
