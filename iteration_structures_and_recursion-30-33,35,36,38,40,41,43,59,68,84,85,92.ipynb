```python
# Q 31. Write a Python program to find Strong Number
def fact(n):
  p = 1
  for i in range(2, n + 1):
    p *= i
  return p


n = int(input("Enter a positive integer: "))
original = n
s = 0
while n != 0:
  d = n % 10
  s += fact(d)
  n //= 10

if s == original:
  print(original, "is a Strong Number")
else:
  print(original, "is not a Strong Number")
```

```python
# 43) Write a program to find whether a no. is Armstrong number:

# Hint : 153=1**3+5**3+3**3

import math
n = int(input("Enter a positive number : "))
original = n
s = 0
nod = math.floor(math.log(n, 10)) + 1
while n != 0:
  d = n % 10
  s += d**nod
  n //= 10

if s == original:
  print(original, "is an Armstrong number")
else:
  print(original, "is not an Armstrong number")
```
