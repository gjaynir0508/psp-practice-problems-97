```python
# Q 2. Illustrate String Literals and escape sequences
# Strings in Python are a sequence of characters. Strings are immutable.
# Literals are notations for constant values of some built-in types.
# String Literals can be enclosed in either single, double or triple quotes (', ", ''' or """)
s = 'string'
a = "hello"
# Triple quoted string literals can span across multiple lines unline the ' or "
c = '''world'''
d = """this 
is a multiline string 
which is enclosed in triple quotes"""

# Concatenation String Literals
e = ("string "
     "literal "
     "concatenation"
     )                  # Equivalent to "string " + "literal " + "concatenatioin"

print(s, a, c, d, e, "\n")

# Escape sequences
# \<newline>    -  Backslash and newline ignored
print("hello \
world")
# \\            -  Backslash
print("backslash\\")
# \'            -  Single Quote
print('single quote\'')
# \"            -  Double Quote
print("double quote\"")
# \a            -  ASCII Bell (BEL)
print("ASCII Bell \a")
# \b            -  ASCII Backspace (BS)
print("ASCII backspace\b after it")
# \f            -  ASCII Formfeed (FF)
print("ASCII line feed \f after line feed")
# \n            -  ASCII Linefeed or Newline (LF)
print("New line \n after new line")
# \r            -  ASCII Carriage Return (CR)
print("Carriage Return\r012345")
# \t            -  ASCII Horizontal Tab (TAB)
print("Horizontal Tab\tafter Tab")
# \v            -  ASCII Vertical Tab (VT)
print("Vertical Tab\v")
# \ooo          -  Character with octal value `ooo`
print("Character with octal value `\\110\\145\\154\\154\\157` - \110\145\154\154\157")
# \xhh          -  Character with hex value `hh`
print("Character with hex value `\\x48\\x65\\x6c\\x6c\\x6f` - \x48\x65\x6c\x6c\x6f")

# Ones that are recognized only in string literals
# \N{name}      -  Character named `name` in the Unicode database
print("Named\N{BS}")  # Prints BackSlash after Named -> Name
# \uxxxx        -  Character with 16-bit hex value `xxxx`
# Prints BackSlash after Named -> Name
print("16-bit Hex `\\u0e4f\\u032f\\u0361\\u0e4f` - \u0e4f\u032f\u0361\u0e4f")
# \Uxxxxxxxx    -  Character with 32-bit hex value `xxxxxxxx`
# Prints BackSlash after Named -> Name
print("32-bit Hex \U00000e4f\U0000032f\U00000361\U00000e4f")
```

```python
# Q 14. Illustrate String: Indexing, slicing, immutablility, predefined operations
# String: In Python, a string is a sequence of characters (of character set i.e., ASCII and Unicode)
# string are enclosed in single, double or triple quotes
s = 'hello'

# Indexing: Each character in a string can be accessed by its position called index
# Index of any character is a number and numbering starts from 0 and goes till n - 1 where n is the length (no. of characters)
# Indexing can be done by subscripting using x[p] where x is sequence and p is index
print(s[0])  # Output: h

# Negative indices can be used to access elements from end
#  0  1  2  3  4
#  h  e  l  l  o
# -5 -4 -3 -2 -1
print(s[-2])  # Output: l

# Slicing: We can get a part (sub-string) of the sequence by slicing it
# Method 1: using slice(start, stop, step)
# start is inclusive, stop is exclusive
sub = slice(0, 4, 2)
print(s[sub])  # Output: hl

# Method 2: using array slice [start:stop:step]
print(s[-4:-1])  # Output: ell
# Output: hello  # default value of start -> 0, stop -> len(s) - 1, step -> 1
print(s[::])
print(s[::-1])  # Output: olleh # reversed version of string

# Immutability: Strings are immutable i.e., once created, their value can't be changed
# there are no mutating methods on a string.
# Even concatenation and multiplication result in new strings while the actual string is left unchanged
# This can be identified by looking at id() value. Different id value means a new string
print(id(s))
s = s + " world"
print(id(s))

# Predefined Operations
# i. Concatenation and multiplication
s_10 = s * 10
print(s_10)

# ii. .title()  -> Capitalizes all the first letters of each word
s_t = s.title()
print(s_t)

# iii. .swapcase()  -> Swaps the case of all characters of string
s_s = s.swapcase()
print(s_s)

# iv. String Formatting using .format()
s_form = "{} {} is {}".format("Python", 3.11, "fun")
print(s_form)

# v. .join()  -> Joins the iterable with the help of the string
s_joined = " - ".join(["Cat", "Bat", "Dog", "Bag"])
print(s_joined)
```

```python
# 58) Write a program to find whether a string is palindrome:

n = input()
m = n[::-1]
print(m)
if m == n:
  print("Palindrome")
else:
  print("Not Palindrome")

a = input()
b = a[::-1]
print(b)

if b == a:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# Q 94. Write a program in python to determine the unique words and their frequency in a given string
# NOTE: To split the string at all delimiters after spaces, we use nested list comprehension
string = input("Enter the sentence / string:\n")
words = [z for x in string.lower().split(" ") for y in x.split("-")
         for z in y.split(".") if z.strip() != ""]
visit_track = [0] * len(words)
frequency = {}
for i in range(len(words)):
  if visit_track[i] == -1:
    continue
  else:
    word_count = 0
    cur_word = words[i]
    # Logic to find no. of instances of the same word in a later part. You can even use words.count(...)
    for j in range(i, len(words)):
      if words[j] == cur_word:
        word_count += 1
        visit_track[j] = -1
    frequency[words[i]] = frequency.setdefault(words[i], 0) + word_count
    visit_track[i] = -1

print("----------------------------")
print("|     Word     | Frequency |")
print("----------------------------")
for word, count in frequency.items():
  print(f"| {word:<13}|{count:^11}|")
print("----------------------------")
```
